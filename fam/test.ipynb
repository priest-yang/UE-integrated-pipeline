```python
import os
import sys
import csv
from collections import deque
from time import time

cur_dir = os.path.dirname(os.path.abspath("__file__"))  # Gets the current notebook directory
src_dir = os.path.join(cur_dir, '../')  # Constructs the path to the 'src' directory
# Add the 'src' directory to sys.path
if src_dir not in sys.path:
    sys.path.append(src_dir)
    
from fam.FiniteAutomationMachine import *
```

```python
file_path = "/home/shaoze/Documents/Boeing/UE-integrated-pipeline/data/PandasData/Sampled_state/PID001_NSL.pkl"
file_buffer = deque(maxlen=40)

model = FiniteAutomationMachine()
time_list = []

df = pd.read_pickle(file_path)

for i, row in df.iterrows():
    
    file_buffer.append(row)
    # test the buffer at last
    if len(file_buffer) == file_buffer.maxlen:
        # run the model
        states = []
        start_time = time()
        for i in file_buffer:
            if model.current_state is None:
                model.run(features=i, current_state=ErrorState)
            else:
                model.run(features=i)
            states.append(model.current_state.name)

        end_time = time()
        time_list.append(end_time - start_time)
        print("Time elapsed: ", end_time - start_time)
        print(states)


print("\n\n\n")
print("Elapsed time: ", sum(time_list))
print("Processed ", len(time_list), " groups.")
print("Speed: ", len(time_list) / sum(time_list))
    
```
